```python
NAME_NER_LABELS = ["PERSON"]
import spacy
import re
import sys
sys.path.append("../") # Adds higher directory to python modules path.
sys.path.append("../../")
from ModelUtils import common, ner_util
#spacy.cli.download('en_core_web_lg')
nlp = spacy.load('en_core_web_lg')
text= "apple inc. is owned by kennedy, j 121 "
doc = nlp(text)
print([(e.text, e.label_, e.start, e.end) for e in doc.ents])

entity_result = []
full_names = [[]]

for token in doc:
    # for att in dir(token):
    #     print (att, getattr(token,att))


    row =   {
        'begin': token.idx,
        'token': token.text,
        'label': token.ent_type_,
        'confidence': None
    }

    if(row['label'] in NAME_NER_LABELS or (len(full_names[-1]) > 0 and re.match('^([\s,])+$', row['token']))):
            #group names appearing sequentially separated by spaces/comma into full names
            full_names[-1].append(row)
    else:
        #start a new full name array
        if len(full_names[-1]) > 0:
            full_names.append([])
                       
for full_name in full_names:
    ner_util.label_name_parts(full_name, entity_result, NAME_NER_LABELS)
            

print("full_names",full_names)
print("entity_result",entity_result)
```
